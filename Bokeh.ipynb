```python
Q-1:
# Import necessary modules
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 8]

# Create a figure
p = figure(title="Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a circle glyph
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Show the plot
output_notebook()  # Use this line if you are working in a Jupyter Notebook
show(p)
```

```python
Q-3:
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare some data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 8]
y2 = [8, 4, 6, 2, 7]

# Create a figure
plot = figure(title="Customized Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis', width=500, height=400)

# Add line glyphs
line1 = plot.line(x, y1, line_width=2, line_color="blue", legend_label="Line 1")
line2 = plot.line(x, y2, line_width=2, line_color="orange", legend_label="Line 2")

# Customize axes
plot.xaxis.axis_label_text_color = "green"
plot.yaxis.axis_label_text_color = "red"
plot.xaxis.major_label_text_color = "purple"
plot.yaxis.major_label_text_color = "brown"

# Customize title
plot.title.text_color = "navy"
plot.title.text_font_style = "italic"
plot.title.text_font_size = "16pt"

# Customize legend
plot.legend.label_text_color = "darkcyan"
plot.legend.label_text_font_style = "bold"
plot.legend.label_text_font_size = "12pt"

# Show the plot
output_notebook()
show(plot)
```

```python
Q-4:
from bokeh.io import curdoc
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.layouts import layout
from bokeh.models.widgets import Slider

# Create initial data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 8]
source = ColumnDataSource(data=dict(x=x, y=y))

# Create a plot
plot = figure(plot_height=300, plot_width=600, title="Interactive Plot", tools="pan,box_zoom,reset")

# Add a line glyph
line = plot.line('x', 'y', source=source, line_width=2, line_alpha=0.8)

# Define a callback function for updating the plot based on a slider
def update(attr, old, new):
    # Update data based on slider value
    y = [i * slider.value for i in x]
    source.data = dict(x=x, y=y)

# Create a slider widget
slider = Slider(start=1, end=5, value=1, step=1, title="Multiplier")
slider.on_change('value', update)

# Create a layout for the plot and slider
layout = layout([[slider], [plot]])

# Add the layout to the current document
curdoc().add_root(layout)
bokeh serve --show interactive_plot.py
```
